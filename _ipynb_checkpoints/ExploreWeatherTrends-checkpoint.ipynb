##Exploring Weather Trends

```python
#Importing needed libraries for analysis
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Importing data files
#Global Average Temperature data by year
global_temp = pd.read_csv(r'C:\Users\ZoilaFlores\OneDrive\Documents\Programming\data-analyst\weather_analysis\global_results.csv')

#City Average Temperature data by year
city_temp = pd.read_csv(r'C:\Users\ZoilaFlores\OneDrive\Documents\Programming\data-analyst\weather_analysis\New_York_results.csv')
```

```python
#Initial look at global average temperature data
plt.plot(global_temp['year'],global_temp['avg_temp'],label='Global',color='lightcoral')
plt.legend()
plt.xlabel("Year")
plt.ylabel("Temperature (Â°C)") 
plt.title("Global Average Temperature")
plt.show()
```

```python
#Initial look at New York City average temperature data
plt.plot(city_temp['year'],city_temp['avg_temp'],label='New York City', color='#4b0082')
plt.legend()
plt.xlabel("Year")
plt.ylabel("Temperature (Â°C)") 
plt.title("New York City Average Temperature")
plt.show()
```

```python
glb_mv_avg = global_temp['avg_temp'].rolling(50).mean()
local_mv_avg = city_temp['avg_temp'].rolling(50).mean()
```

```python
#Local Data is as same as Cairo
plt.plot(global_temp['year'],glb_mv_avg,label='Global')
plt.plot(city_temp['year'],local_mv_avg,label='New York City')
plt.legend()
plt.xlabel("Years")
plt.ylabel("Temperature (Â°C)") 
plt.title("New York City Average Temperature")
plt.show()
```
